##Splitting According to Color

D878677F-AC5A-4F00-99B5-890E9589912D_1_201_a.jpeg

```python
#Total Entropy
TEntropy = round((-3/7*log2(3/7)-4/7*log2(4/7)),3)
#Color Entropy
BrownEnt = round((-1/3*log2(1/3)-2/3*log2(2/3)),3)
WhiteEnt = round((-2/4*log2(2/4)-2/4*log2(2/4)),3)
InfoGainColor = round((TEntropy - 3/7 * BrownEnt - 4/7 * WhiteEnt),3)
print(InfoGainColor)
```

##Splitting According to Flight

189A8C6A-0101-4E71-A253-256F5C5CB86F_1_201_a.jpeg

```python
#Fly Entropy
FlyYesEnt = round((-3/3*log2(3/3)),3)
print(FlyYesEnt)
FlyNoEnt = round((-3/4*log2(3/4) - 1/4*log2(1/4)),3)
print(FlyNoEnt)
InfoGainFly = round((TEntropy - 3/7*FlyYesEnt - 4/7*FlyNoEnt),3)
print(InfoGainFly)
```
